```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\shano\Downloads\Diwali Sales Data.csv", encoding='ISO-8859-1')
df.head(5)
```

```python
df.rename(columns={"Marital_Status":"Marriage"},inplace=True)
df.head(5)
```

```python
df.dtypes
df.isnull().sum()
df.dropna(how = "all" , axis =1 , inplace=True)
df.dropna(how = "any" , axis =0 , inplace=True)
df["Amount"]=df["Amount"].astype('int')
df.head()
```

```python
df["Amount"].describe()
```

Gender 

```python
df.columns
ax = sns.countplot(x="Gender",data=df,palette="flare" )
plt.figure(figsize=(0.5,0.3))
for p in ax.containers:
    ax.bar_label(p)
```

```python
ga=df.groupby(["Gender"],as_index=False)["Amount"].sum().sort_values(by= "Amount" , ascending=False)
ga
```

```python
ax = sns.barplot(x="Gender",y="Amount",data=ga,palette="YlGnBu")
plt.figure(figsize=(10,5))
```

From above graphs we can tell that majority of the buyers were females(7832) and had more Purchasing power than men (74335853).

AGE

```python
df.columns
ao = sns.countplot(x="Age Group",data = df,hue = "Gender")
for i in ao.containers:
    ao.bar_label(i)
```

```python
aa=df.groupby(["Age Group"],as_index=False)["Amount"].sum().sort_values(by= "Amount" , ascending=False)
aa
```

```python
al =sns.barplot(x="Age Group",y="Amount",data = aa,palette="magma")
```

From above graphs,majority of the buyers are females between age 26-35.

State

```python
sam=df.groupby(["State"],as_index=False)['Amount'].sum().sort_values(by="Amount",ascending =False).head(10)
plt.figure(figsize=(16,5))
sam_plot =sns.barplot(x='State',y="Amount",data = sam,palette='viridis')
```

From above graph,majority of the purchasers were from Uttarpradesh,Maharashtra and Karnataka

Occupation

```python
oamt = df.groupby(["Occupation"],as_index=False)["Amount"].sum().sort_values(by=["Amount"],ascending=False).head(10)
plt.figure(figsize=(16,5))
omt_plot=sns.barplot(x="Occupation",y="Amount",data=oamt,palette="plasma")
```

From above graph majority of purchasers were from IT sector,Health care and aviation.
